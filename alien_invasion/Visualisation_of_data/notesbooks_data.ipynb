```python
import matplotlib.pyplot as plt
```

```python
# Tracer une courbe simple
#mpl_squares.py
squares = [1, 4, 9, 16, 25]
fig, ax = plt.subplots()
ax.plot(squares)

plt.show()
```

```python
# Modifier le type de libell et l'paisseur des lgines

#mpl_squares.py
squares = [1, 4, 9, 16, 25]
fig, ax = plt.subplots()
ax.plot(squares, linewidth=3)

#Fixer le titre du graphique et les libelles des axes.
ax.set_title("Nombres carres",fontsize=24)
ax.set_xlabel("Valeur",fontsize=14)
ax.set_ylabel("Carr de la valeur",fontsize=14)

#Fixer la taille des graduations.
ax.tick_params(axis='both',labelsize=14)

plt.show()
```
